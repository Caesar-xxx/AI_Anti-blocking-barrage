```python
# 导入常见包
import cv2 
import numpy as np
import matplotlib.pyplot as plt 
plt.rcParams['figure.dpi'] = 200
```

```python
# 读取图片
img = cv2.imread('./images/street.jpg')
img_cvt = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_cvt)
```

```python
# 导入实例分割得包
# 导入pixellib
import pixellib
```

```python
# 导入实例分割
from pixellib.instance import instance_segmentation
```

```python
# 实例化
instance = instance_segmentation()
# 加载模型
instance.load_model('./weights/mask_rcnn_coco.h5')
```

```python
# 实例分割
target_classes = instance.select_target_classes(car=True)
results,output = instance.segmentFrame(img.copy(),segment_target_classes=target_classes)
```

```python
plt.imshow(cv2.cvtColor(output,cv2.COLOR_BGR2RGB))
```
